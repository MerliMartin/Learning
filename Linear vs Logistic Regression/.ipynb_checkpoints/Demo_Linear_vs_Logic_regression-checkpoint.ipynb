```python
import pandas as pd  # for a dataframe, it's just a way of storing the data

from sklearn.preprocessing import LabelEncoder  # pre-processing labeling coder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, confusion_matrix  # how accurate is our data
```

```python
dataset = pd.read_csv('Datasets/insurance.csv')
dataset
```

```python
 dataset.info()
```

```python
# Changing values sex, smoker and region to numeric values (0 and 1)
label = LabelEncoder()

# sex: male or female (female=0 and male=1)
label.fit(dataset.sex.drop_duplicates())
dataset.sex = label.transform(dataset.sex)

# smoker or not (yes=1 and no=0)
label.fit(dataset.smoker.drop_duplicates())
dataset.smoker = label.transform(dataset.smoker)

# region: southwest(3) or southeast(2) or northwest(1)
label.fit(dataset.region.drop_duplicates())
dataset.region = label.transform(dataset.region)

dataset
```

```python
X_lin = dataset.drop(['charges'], axis=1)
y_lin = dataset[['charges']]

# we don't want to do all the data at once
# we are going to take 30% for testing and 70% for training
X_lin_train, X_lin_test, y_lin_train, y_lin_test = train_test_split(X_lin, y_lin, test_size=0.3, random_state=42)

Linear_model = LinearRegression()
Linear_model.fit(X_lin_train, y_lin_train)
```

```python
for idx, col_name in enumerate(X_lin_train.columns):
    print(f"The coefficient for {col_name} is {Linear_model.coef_[0][idx]}")
```

```python
# fun to look - it starts at this negative number and then adds all the numbers to it
# that's all that means that's our intercept on there 
# intercept - number punkt, mis lbib y-telge , ehk kui x=0
intercept = Linear_model.intercept_[0]
intercept
```

```python
y_pred = Linear_model.predict(X_lin_test)

pred = pd.DataFrame(X_lin_test)
pred['charges'] = y_lin_test
pred['pred_charges'] = y_pred
pred
```

```python
mushroom_data = pd.read_csv('Datasets/mushrooms.csv')
mushroom_data
```

```python
# The data is categorical, converting to numberical
labelencoder = LabelEncoder()

for col in mushroom_data.columns:
    mushroom_data[col] = labelencoder.fit_transform(mushroom_data[col])
    
mushroom_data
```

```python
X_log = mushroom_data.drop(['type'], axis=1)
y_log = mushroom_data[['type']]

X_log_train, X_log_test, y_log_train, y_log_test = train_test_split(X_log, y_log, test_size=0.3, random_state=42)

Logistic_model = LogisticRegression(max_iter=500)
Logistic_model.fit(X_log_train, y_log_train)  # fitting only 70% of data
```

```python
pred = Logistic_model.predict(X_log_test) # predicting from 30% of given data
```

```python
Logistic_model.score(X_log_test,y_log_test)
```

- pretty accurate , that's good 

```python
confusion_matrix(y_log_test, pred)
```

- poisonous was 1 and edible was 0
- we have True, True, True, False / True, False, False, False
- out of the edible mushrooms we correctly labeled 1201 mushrooms edible, that were edible
- and we correctly measured 1113 poisonous  mushrooms as poisonous

- we labeled 56 edible mushrooms as being poisonous
- we measured 68 mushrooms as being edible that were poisonous

! Not the Best model to predict whether you're going to eat mushroom or not and you want to dig a little deeper before you start picking mushrooms off the side of the trail
